```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib 
import os
from glob import glob

root = "./data/"
```

## 1. Load

```python
data = {}
for file in os.listdir(root):
    data[file.split("_")[-1].split(".")[0]] = pd.read_csv(os.path.join(root, file))
```

```python
data.keys()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
sns.countplot(data["AnomalyLabels"], x="Label", ax=ax[0])
sns.countplot(data["StateMachineLabel"], x="Label", ax=ax[1])
plt.show()
```
